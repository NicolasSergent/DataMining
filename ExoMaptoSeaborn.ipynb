```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
# 1° question total des bills selon le sex
sns.barplot(x="sex",y="total_bill",hue="sex",data=tips)
plt.draw()
```

```python
# 2°  question distribution par rapport au total des bills
sns.distplot(tips["total_bill"])
plt.draw()
```

```python
# 3° question jointplot
sns.jointplot(x="total_bill", y="tip", data=tips)
plt.draw()
```

```python
# faire une regression linéaire
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
# 4° question fonction pairplot 
sns.pairplot(data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
import pandas
```

```python
google_app= pandas.read_table("googleplaystore.csv",sep=",")
google_app
```

```python
google_app_sans_na = google_app.dropna()
```

```python
google_app_sans_na["Price"]=google_app_sans_na["Price"].apply(lambda x: x.replace("$" , ""))
google_app_sans_na["Price"]=google_app_sans_na["Price"].apply(lambda x: float(x))
```

```python
google_app_sans_na["Reviews"]=google_app_sans_na["Reviews"].apply(lambda x: int(x))
```

```python
google_app_sans_na["Installs"]=google_app_sans_na["Installs"].apply(lambda x: x.replace("," , ""))
google_app_sans_na["Installs"]=google_app_sans_na["Installs"].apply(lambda x: x.replace("+" , ""))
```

```python
google_app_sans_na["Installs"]=google_app_sans_na["Installs"].apply(lambda x: int(x))
```

```python
google_app_sans_na
```

```python
#question 1 noms,categories et genre des apply les plus installes
apply_max_installer=max(google_app_sans_na["Installs"])
google_app_sans_na.loc[(google_app_sans_na["Installs"]==apply_max_installer),["App","Category","Genre"]].head()
```
